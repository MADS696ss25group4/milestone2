```python
import pandas as pd
```

```python
# Read in hurricane data from HurDat2 file
filename = 'hurdat2-1851-2024-040425.txt'
with open(filename) as file:
    lines = [line.split(',') for line in file]

# Initiate empty dataframe
hurricanes_df = pd.DataFrame(columns=['Code', 'Name', 'Date',
                                      'Time in UTC', 'Record Identifier',
                                      'Status','Latitude','Longitude',
                                      'Max Wind','Min Pressure',
                                      '34 kt wind radii maximum extent in northeastern quadrant',
                                      '34 kt wind radii maximum extent in southeastern quadrant',
                                      '34 kt wind radii maximum extent in northwestern quadrant',
                                      '34 kt wind radii maximum extent in southwestern quadrant',
                                      '50 kt wind radii maximum extent in northeastern quadrant',
                                      '50 kt wind radii maximum extent in southeastern quadrant',
                                      '50 kt wind radii maximum extent in northwestern quadrant',
                                      '50 kt wind radii maximum extent in southwestern quadrant',
                                      '64 kt wind radii maximum extent in northeastern quadrant',
                                      '64 kt wind radii maximum extent in southeastern quadrant',
                                      '64 kt wind radii maximum extent in northwestern quadrant',
                                      '64 kt wind radii maximum extent in southwestern quadrant',
                                      'Radius of Max Wind'])


# Initiate a counter
row_counter = 0
next_storm_start_line = 0

while row_counter < len(lines):

  if row_counter == len(lines)-1:
      break

  storm_code = lines[row_counter][0]
  storm_name = lines[row_counter][1].strip()
  next_storm_start_line = int(lines[row_counter][2]) + next_storm_start_line + 1


  while row_counter < next_storm_start_line:


    if row_counter == len(lines)-1:
      break
    else:
      row_counter += 1

    if (lines[row_counter][3].strip()=='HU')|(lines[row_counter][3].strip()=='TS')|(lines[row_counter][3].strip()=='TD'): #and (lines[row_counter][2].strip() == 'L'):
      #add Puerto Rico storms to dataset
        if (float(lines[row_counter][4][:-1].strip())<19.0)&(float(lines[row_counter][4][:-1].strip())>17.0):
            if(float('-'+lines[row_counter][5][:-1].strip())>-68.0)&(float('-'+lines[row_counter][5][:-1].strip())<-65.0):
              
                hur_data_list = [] # Initiate an empty list

                for i in range(len(lines[row_counter])):
                    hur_data_list = hur_data_list + [lines[row_counter][i].rstrip()]
              
                hurricanes_df.loc[len(hurricanes_df)] = [storm_code, storm_name]+hur_data_list

              #row_counter = next_storm_start_line # Ensures only first landfall event is captured
      #check if latitude is south of 24.4N, if so, skip to next observation
        elif (float(lines[row_counter][4][:-1].strip())<24.4):  
            continue
      #filter out landfalls in the Bahamas
        elif (float(lines[row_counter][4][:-1].strip())<27.5)&(float(lines[row_counter][4][:-1].strip())>18.5)&(float('-'+lines[row_counter][5][:-1].strip())>-79.3):
            continue
        else:
            hur_data_list = [] # Initiate an empty list

            for i in range(len(lines[row_counter])):
                hur_data_list = hur_data_list + [lines[row_counter][i].rstrip()]

            hurricanes_df.loc[len(hurricanes_df)] = [storm_code, storm_name]+hur_data_list

          #row_counter = next_storm_start_line # Ensures only first landfall event is captured
```
